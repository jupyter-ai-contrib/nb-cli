# Large Notebook Sanity Test

This notebook exercises execution across many cells.

```python
data = list(range(100))
total = sum(data)
print(f"total={total}")
```

## Section 1: Computation

```python
results = []
for i in range(5):
    val = math.sqrt(i * 100)
    results.append(val)
    print(f"sqrt({i*100})={val:.2f}")
```

```python
mapping = {f"key_{i}": i**2 for i in range(10)}
print(f"keys={len(mapping)}")
```

```python
text = "hello world " * 50
words = text.split()
print(f"word_count={len(words)}")
```

## Section 2: State Propagation

```python
combined = total + len(mapping) + len(words)
print(f"combined={combined}")
```

```python
nested = {"data": data[:5], "results": results, "mapping_subset": {k: v for k, v in list(mapping.items())[:3]}}
print(json.dumps(nested, indent=2))
```

```python
class Counter:
    def __init__(self, start=0):
        self.value = start
    def increment(self, by=1):
        self.value += by
        return self.value
    def __repr__(self):
        return f"Counter({self.value})"

c = Counter(10)
print(c)
```

```python
for _ in range(5):
    c.increment(3)
print(f"counter_final={c.value}")
```

## Section 3: Multi-line Output

```python
for i in range(20):
    print(f"line_{i:03d}: {'*' * (i+1)}")
```

```python
try:
    result = 100 / 4
    print(f"safe_division={result}")
except ZeroDivisionError:
    print("ERROR: division by zero")
```

```python
table = []
for row in range(5):
    cols = [f"{row},{col}" for col in range(4)]
    table.append(" | ".join(cols))
for line in table:
    print(line)
```

## Section 4: Final Assertions

```python
assert total == 4950, f"Expected 4950, got {total}"
assert len(results) == 5
assert c.value == 25, f"Expected 25, got {c.value}"
assert combined == 5060
print("all_assertions_passed=True")
```

```python
summary = {
    "total": total,
    "results_count": len(results),
    "counter": c.value,
    "combined": combined,
    "word_count": len(words),
}
print(f"SUMMARY: {json.dumps(summary)}")
```
