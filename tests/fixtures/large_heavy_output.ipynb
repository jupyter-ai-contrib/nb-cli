```python
import json
data = [{"id": i, "value": i**2, "label": f"item_{i}"} for i in range(50)]
print(json.dumps(data))
```

```python
header = "| {:>5} | {:>10} | {:>15} |".format("ID", "Value", "Squared")
sep = "|" + "-"*7 + "|" + "-"*12 + "|" + "-"*17 + "|"
print(header)
print(sep)
for i in range(30):
    print("| {:>5} | {:>10} | {:>15} |".format(i, i*3, i**2))
```

```python
matrix = []
for i in range(10):
    row = []
    for j in range(10):
        row.append(i * j)
    matrix.append(row)
    print(f"row_{i}: {row}")
```

```python
words = ["alpha", "beta", "gamma", "delta", "epsilon"]
result = []
for i in range(20):
    w = words[i % len(words)]
    result.append(f"{w}_{i:03d}")
print(" ".join(result))
```
