```
# Cell 0: Import and define
import math
pi_value = math.pi
print(f'π = {pi_value}')
```

```
# Cell 1: Use variable from Cell 0
radius = 5
area = pi_value * radius ** 2
print(f'Area of circle with r={radius}: {area}')
```

```
# Cell 2: Use variables from both cells
perimeter = 2 * pi_value * radius
print(f'Perimeter: {perimeter}')
print(f'Area/Perimeter ratio: {area/perimeter}')
```
